# Encode Categorical data to numerical

```python
df = df.drop(['id', 'Berat Akhir', 'Jmlh Tahun Pendidikan', 'Keuntungan Kapital', 'Kerugian Capital'], axis=1)
```

```python
df
```

```python
# We have categorical data are Kelas Pekerja, Pendidikan, Status Perkawinan, Pekerjaan, Jenis Kelamin and Gaji

# Encode categorical data using Label Encoder

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['Pendidikan'] = le.fit_transform(df['Pendidikan'])
df['Jenis Kelamin'] = le.fit_transform(df['Jenis Kelamin'])
df['Gaji'] = le.fit_transform(df['Gaji'])
```

```python
df
```

```python
df = pd.get_dummies(data=df, columns=['Kelas Pekerja', 'Status Perkawinan', 'Pekerjaan'])
```
